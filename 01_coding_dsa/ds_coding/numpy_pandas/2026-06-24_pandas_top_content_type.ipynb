## Problem
For **each user**, find the **single `content_type` they spent the most total watch time on**, and report that total.

**Output:** one row per user with columns `user_id`, `top_content_type`, `total_secs`.

Run the setup cell, then write your solution in the answer cell. Run it and check against *Expected output* at the bottom. When done, tell me in chat and I'll review correctness + whether it's idiomatic.

````python
import pandas as pd

df = pd.DataFrame({
    'user_id':      [1, 1, 1, 2, 2, 3, 3, 3, 3],
    'content_id':   ['A','B','A','C','A','B','B','C','A'],
    'watch_secs':   [120, 300, 45, 600, 30, 500, 200, 50, 410],
    'content_type': ['movie','show','movie','movie','movie','show','show','movie','movie'],
})
df
````

## Your solution
Write the pandas code below. Aim for vectorized / idiomatic (groupby chains, no Python loops over rows).

````python
totals = df.groupby(['user_id' , 'content_type'])['watch_secs'].sum()
totals
````

````python
totals
````

````python
win_rows = totals.groupby(['user_id'])['watch_secs'].idxmax()
````

````python
win_rows
````

````python
totals.loc[win_rows]
````
